# investors column data clean

```python
###importing libraries
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("startup_funding_Data _kaggle.csv")
df
```

```python
df_new = df.iloc[:,6:7]
df_new
```

```python
df_new
```

```python
# not including colums names with undisclosed investors
df_new = df_new[df_new["Investors Name"] != "Undisclosed Investors"]
df_new = df_new[df_new["Investors Name"] != "Undisclosed investors"]
```

```python
df_new["Investors Name"].value_counts() # checking that undsclosed investors have removed
```

```python
arr1 = []
for i in df_new["Investors Name"].value_counts().index:
    arr1.append(i)
    
arr2 = []
for i1 in df_new["Investors Name"].value_counts().values:
    arr2.append(i1)
    
arr1 = arr1[0:5]
arr2 = arr2[0:5]
```

```python
# plotting top 5 investors on the pie chart 
i = 0
while i < len(arr1):
    print(arr1[i],"--->", arr2[i])
    i = i + 1


plt.pie(arr2, labels = arr1,autopct='%0.2f%%',counterclock=False,startangle=110)
plt.title('Top 5 Investors')
plt.show()
```
