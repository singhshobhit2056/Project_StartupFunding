```python
df = df.iloc[:,5:9] # selecting usefull columns
df
```

```python
df = df.drop(columns = ["Investors Name", "InvestmentnType"]) # dropping multiple useless columns
```

```python
df = df.dropna() # dropping na values
```

```python
df["City  Location"].value_counts()
```

```python
df["City  Location"].unique()
```

```python
def seprate(x):
    return x.split("/")[0].strip()

df["City  Location"] = df["City  Location"].apply(seprate) # This funtion seprates the multipe city (continue reading next line)
#location with "/" symbol and also prints the first word with no spaces(.strip() removes the spaces between them.)
df
```

```python
df["City  Location"].unique() # we can clearly identify the removed slashed elements from City  Location column.
```

```python
# identified some similar names and so cleaning then before to get good results
df["City  Location"].replace("Delhi", "New Delhi", inplace = True)
df["City  Location"].replace("bangalore","Bangalore", inplace = True)
```

```python
df["City  Location"].replace("Bengaluru","Bangalore", inplace = True)
```

```python
df["City  Location"].value_counts()
```

```python
df["Amount in USD"].unique() #checking different type of values in the amount
```

```python
#handling unkown data which can be seen above replacing them with nan 
arr = []
for i in df["Amount in USD"]:
    arr.append(i)
```

```python
import numpy as np
arr2 = []
for i in arr:
    if "x" in i:
        arr2.append(np.nan)
        continue
    elif "+" in i:
        arr2.append(np.nan)
        continue
    elif "," in i:
        arr2.append(i)
        continue
    else :
        arr2.append(np.nan)
    
```

```python
print(arr2)
```

```python
df
```

```python
df["amount"] = arr2 # added a new updated amount with removed all type of unwanted things in data
```

```python
df = df.drop("Amount in USD", axis = 1)
```

```python
df #checking new dataframe
```

```python
df['amount'] = df['amount'].apply(lambda x: float(str(x).replace(",","")))
```

```python
df #checking the updated data frame
```

```python
df
```

```python
df["City  Location"].value_counts()
```

```python
df_new = df.groupby('City  Location').sum(["amount"]) # adding all the amount for the same cities
```

```python
df_new = df_new.sort_values(by = "amount", ascending = False) # sorting the added values 
```

```python
df_new # seeing the results
```

```python
array1 = [] # getting array of top 5 cities with max sum of amount
for i in range (0,5):
    array1.append(df_new.index[i])
array1
```

```python
array2 = [] # getting array of amount of above top cities
for i1 in range (0,5):
    array2.append(df_new.values[i1][0])
array2
```

```python
import matplotlib.pyplot as plt #plotting in bar graph top cities with their sum of prices in cities
i = 0
while i < len(array1):
    print(array1[i],"--->", array2[i])
    i = i + 1


plt.bar(array1, array2, color = ['red', 'green', 'blue', 'orange', 'pink'])
plt.ylabel('TOTAL FUNDS IN A CITY(x10 power 10 in dollars)')
plt.xlabel('CITY')
plt.title('CITY vs TOTAL FUNDS IN A CITY')
plt.show()
```
