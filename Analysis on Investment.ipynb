# Analysis on type of Investment and their share in the investment amount

```python
###importing libraries
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("startup_funding_Data _kaggle.csv")
df
```

```python
df_new = df.iloc[:,7:9]
df1 = df_new.iloc[:,0:1]
```

```python
df1
```

```python
df1.columns
```

```python
df1.dropna()
```

```python
df1["InvestmentnType"].value_counts() #Seed/ Angel Funding
```

```python
# cleaning the column 
df1["InvestmentnType"].unique()# finding unique elements in the data
```

```python
df1["InvestmentnType"].replace("pre-series A", "Pre-series A", inplace = True)
df1["InvestmentnType"].replace("Angel / Seed Funding", "Seed/ Angel Funding", inplace = True)
df1["InvestmentnType"].replace("Seed / Angle Funding", "Seed/ Angel Funding", inplace = True)
df1["InvestmentnType"].replace("Seed/Angel Funding", "Seed/ Angel Funding", inplace = True)
df1["InvestmentnType"].replace("Seed\\\\nFunding", "Seed/ Angel Funding", inplace = True)
df1["InvestmentnType"].replace("Seed/Angel Funding", "Seed/ Angel Funding", inplace = True)
df1["InvestmentnType"].replace("PrivateEquity", "Private Equity Round", inplace = True)
```

```python
df1["InvestmentnType"].value_counts() # just checking the replaced values 
```

```python
# concidering only top 5 investment types
arr1 = []
for i in df1["InvestmentnType"].value_counts().index:
    arr1.append(i)
    
arr2 = []
for i1 in df1["InvestmentnType"].value_counts().values:
    arr2.append(i1)
arr1 = arr1[0:5]
arr2 = arr2[0:5]
```

```python
print(arr1) # checking that the abbove code is working properly
print(arr2)
```

# pie chart representation for top 5 Investment types

```python
i = 0
while i < len(arr1):
    print(arr1[i],"--->", arr2[i])
    i = i + 1
    
    
plt.pie(arr2,labels=arr1, autopct='%0.2f%%', counterclock=False, startangle=110, explode = [0,0,0,0.5,1.2])

plt.show()
```
